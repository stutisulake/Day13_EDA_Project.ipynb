```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis

df = pd.read_csv("housing.csv")

plt.figure(figsize=(8,5))
sns.histplot(df['Price'], kde=True, bins=30)
plt.title("Histogram of House Prices with KDE")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(8,5))
sns.boxplot(y=df['Price'])
plt.title("Boxplot of House Prices")
plt.ylabel("Price")
plt.show()

price_skewness = skew(df['Price'])
price_kurtosis = kurtosis(df['Price'])

print("Skewness of Price:", price_skewness)
print("Kurtosis of Price:", price_kurtosis)

if price_skewness > 0:
    print("The Price distribution is Positively Skewed (Right-Skewed).")
elif price_skewness < 0:
    print("The Price distribution is Negatively Skewed (Left-Skewed).")
else:
    print("The Price distribution is approximately Normal.")

plt.figure(figsize=(8,5))
sns.countplot(x='City', data=df)
plt.title("Count of Houses by City")
plt.xticks(rotation=45)
plt.show()

print("\nMost Frequent City:")
print(df['City'].value_counts().idxmax())

print("\nUnivariate Analysis Completed Successfully.")

```
